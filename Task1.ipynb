```python
df = pd.read_csv("C://CompletePython//DataScience//BigData//Sales_File1_to_Pandas.csv")
```

```python
df = df.drop(['ProductID', 'ProductName'], axis=1)
```

```python
newdf = df.drop_duplicates(
  subset = ['CustomerID', 'Product_864_Bought'],keep = 'last').reset_index(drop = True)
```
